Exploratory Data Analysis (EDA)

It’s the process of analyzing data sets to summarize their main characteristics—often using visual methods—before applying any modeling or machine learning techniques.Key goals of EDA include:
- Understanding the structure of your data  
- Detecting outliers or anomalies  
- Identifying patterns, trends, or relationships  
- Checking assumptions for statistical modeling

Visualizations

```python
data.plot(kind ='line', title ='Sales Graph')
```

```python
data.plot.area(figsize =(10,5))
```

```python
data.plot.scatter(x ='Units_Sold' , y='Price_Per_Unit' , s = 50 , c ='grey')
```

```python
data.sort_values(by ="Units_Sold",ascending=False).head(5)
```

```python
data.select_dtypes(include='number').corr()
```

```python
sns.heatmap(data.select_dtypes(include='number').corr(), annot=True)
plt.show()
```

```python
data.boxplot(figsize=(5,5))
```

```python
data.select_dtypes(include='number').corr(method= 'pearson') #pearson, kendall, spearman
```

```python
correlation_matrix = data.select_dtypes(include='number').corr(method= 'pearson') 
sns.heatmap(correlation_matrix, annot =True)
plt.title('Sales Correlation Matrix')
plt.show()
```

```python
top_product = data.groupby('Product')['Units_Sold'].sum().sort_values(ascending=False).head(10)
print(top_product)
```

```python
data['Revenue'] = data['Units_Sold'] * data['Price_Per_Unit']
```

```python
data['Revenue'] = data['Units_Sold'] * data['Price_Per_Unit']
top_region = data.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print(top_region)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x='Region', y='Revenue', estimator='sum', ci=None)
plt.title('Total Revenue by Region')
plt.ylabel('Revenue')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=data, x='Price_Per_Unit', y='Revenue')
plt.title('Unit Price vs Revenue')
plt.xlabel('Unit Price')
plt.ylabel('Revenue')
plt.tight_layout()
```
